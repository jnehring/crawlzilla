```python
import fitz

def extract_text_with_paragraphs(pdf_path):
    """
    Extracts text from a PDF, attempting to preserve paragraph structure.
    """
    text_content = []
    try:
        doc = fitz.open(pdf_path)
        for page_num in range(26,27):
            page = doc.load_page(page_num)
            # Extract text blocks, which often correspond to paragraphs
            blocks = page.get_text("blocks")
            
            for block in blocks:
                # block[4] contains the text content
                text_content.append(block[4].strip())
            
            # Add a separator between pages if desired
            if page_num < doc.page_count - 1:
                text_content.append("\n---PAGE BREAK---\n") 
        doc.close()
    except Exception as e:
        print(f"Error processing PDF: {e}")
        return None
    
    return "\n\n".join(text_content) # Join blocks with double newlines for paragraph separation

# Example usage:
pdf_file = "data/CONTRACTS-LAW.pdf" # Replace with your PDF file path
extracted_text = extract_text_with_paragraphs(pdf_file)

if extracted_text:
    print(extracted_text)
```

```python
from bs4 import BeautifulSoup
import re

# Start with an empty HTML document
soup = BeautifulSoup("<html><head></head><body></body></html>", "html.parser")

class HTML2Text:
    """Convert HTML code to text
    """
    def __init__(self):
        self.replace_consecutive_whitespace = re.compile(r'\s+')
        self.nodeTypes = set(["p", "span", "h1", "h2", "h3", "h4", "h5", "h6"])

    def iterate_nodes(self, parent):

        if not "contents" in parent.__dict__.keys():
            return

        for child in parent.contents:
            if child.name in self.nodeTypes:
                yield child
            else:
                for node in self.iterate_nodes(child):
                    yield node

    def clean_text(self, text):

        lines = []
        for line in text.split("\n"):
            line = line.strip()

            if len(line) == 0:
                continue

            if len(line) == 0:
                continue

            # needs to have a minimum length
            if len(line) < 50:
                continue

            # needs to contain at least one sentence marks
            sentence_marks = ".,!?"
            counts = sum([line.count(x) for x in sentence_marks])

            # needs to have a ratio of upper / lower characters
            lower = "abcdefghijklmnobqrstuvwxyz"
            upper = lower.upper()

            lower_ratio = sum(line.count(x) for x in lower) / len(line)
            upper_ratio = sum(line.count(x) for x in upper) / len(line)

            if lower_ratio > 0.95 or upper_ratio > 0.2:
                continue

            # should not end with ...
            needle = "..."
            if line[-3:] == needle:
                continue

            line = self.replace_consecutive_whitespace.sub(" ", line)

            lines.append(line)

        if len(lines) == 0:
            return None
        else:
            lines = list(set(lines))
            return "\n".join(lines)
            
    def extract_text(self, soup, clean_text=True):
        texts = []
        for node in self.iterate_nodes(soup):
            
            if clean_text:
                text = self.clean_text(node.text)
            if text is not None:
                for line in text.split("\n"):
                    texts.append(line)
        return texts


def pdf2html(infile):

    doc = fitz.open(infile)
    soup = BeautifulSoup("<html><head></head><body></body></html>", "html.parser")

    counter=0
    for page in doc:
        counter += 1
        if counter != 44:
            continue
        text = page.get_text("blocks")
        text = [t[4].strip() for t in text]

        segments = []
        for t in text:
            for t2 in t.split("\n\n"):
                t2 = re.sub(r'\n(?![•])', "", t2)
                segments.append(t2)

        for segment in segments:
            new_paragraph = soup.new_tag("p")
            new_paragraph.string = segment
            soup.body.append(new_paragraph)

    return soup

pdf_file = "data/SEZAR-Catalogue.pdf" # Replace with your PDF file path
soup = pdf2html(pdf_file)
html2text = HTML2Text()
result = html2text.extract_text(soup)
print("\n\n".join(result))
```
