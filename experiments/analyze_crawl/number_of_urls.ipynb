```python
from urllib.parse import urlparse
def extract_domains(urllist):
    counts = {}
    for url in urllist:
        domain = urlparse(url).netloc
        if domain in url:
            if domain in counts:
                counts[domain] += 1
            else:
                counts[domain] = 1
    return counts

counts_downloaded = extract_domains(downloaded_urls)
counts_downloaded 
```
