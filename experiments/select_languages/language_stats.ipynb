```python
df = pd.read_csv("~/Downloads/cc_websites.csv.gz", compression="gzip")
counts = pd.DataFrame(df.content_languages.value_counts().reset_index())
counts = {r["content_languages"]: r["count"] for ix, r in counts.iterrows()}
```

```python
df = pd.read_csv("data/languages.csv")
df["n_pages_commoncrawl"] = df.code.map(lambda x:0 if x not in counts.keys() else counts[x])
df = df[['language', 'n_speakers', 'code',
       'script', 'n_pages_commoncrawl']]
df.to_csv("data/languages.csv")
df.sort_values(by="n_pages_commoncrawl", ascending=False)
```
