```python
# data https://www.berlitz.com/blog/what-are-most-spoken-languages-africa

import pandas as pd

x = """3. Swahili
Total speakers: 80 million
swa
Latn
4. Hausa
Total speakers: 70 million
hau
Latn
5. Amharic
Total speakers: 57 million
amh
Ethi
6. Yoruba
Total speakers: 45 million
yor
Latn
7. Oromo
Total speakers: 40 million
orm
Latn
8. Igbo
Total speakers: 40 million
ibo
Latn
10. Zulu
Total speakers: 27 million
zul
Latn
11. Malagasy
Total speakers: 25 million
mlg
Other
12. Fula
Total speakers: 25 million
ful
Latn
13. Somali
Total speakers: 22 million
som
Other
14. Shona
Total speakers: 14 million
sna
Latn
15. Rundi (Kirundi)
Total speakers: 13 million
run
Latn
16. Kinyarwanda
Total speakers: 12 million
kin
Latn
17. Chichewa
Total speakers: 12 million
nya
Latn
18. Akan
Total speakers: 12 million
aka
Latn
19. Tigrinya
Total speakers: 9 million
tir
Other
20. Lingala
Total speakers: 8 million
lin
Latn"""

lines = x.split("\n")
df = []
row = []
for line in lines:
    row.append(line)
    if len(row) == 4:
        df.append(row)
        row = []

for i in range(len(df)):
    df[i][0] = df[i][0][df[i][0].find(" ")+1:].strip()
    df[i][1] = df[i][1].replace("Total speakers: ", "")
    df[i][1] = df[i][1].replace("million", "").strip()
    df[i][2] = df[i][2].strip()

df = pd.DataFrame(df, columns=("language", "n_speakers", "code", "script"))
df.to_csv("data/languages.csv")
df 
```
