```python
import csv
import gzip
import os
import shutil

infile = '/Volumes/PassBlau/download/Nicht bestätigt 105189.crdownload'
outfolder = '/Volumes/PassBlau/download/seeds'

# print(os.path.exists(outfolder))
# if os.path.exists(outfolder):
#     shutil.rmtree(outfolder)

# os.mkdir(outfolder)

headers = None

hashes = set()
writers = {}

i = 0

try:
    with open(infile, 'r', newline='') as csvfile:

        csv_reader = csv.reader(csvfile)
        for row in csv_reader:

            if headers is None:
                headers = row
                continue

            url = row[0]
            h = hash(url)

            if h in hashes:
                continue

            hashes.add(h)

            lang = row[2]

            if lang not in writers.keys():
                writers[lang] = gzip.open(os.path.join(outfolder, lang + ".txt.gz"), 'wt')

            writers[lang].write(url + '\n')

            i += 1
            if i%100000 == 0:
                print(f"{i:,}")
finally:
    for writer in writers.values():
        writer.close()
```
