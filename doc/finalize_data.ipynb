# How to generate the final dataset

This document explains how to generate the final dataset.


You can use the `finalize_data` script to generate the final data files. The script works as follows:

1. It takes as arguments the languages for which you want to create the data.
2. Then, it iterates over all files in the folder `textual_outputs` for this languages. By default, its assumes that the outputs are located in `../outputs`.
3. For each language, it maintains a hash set for each sentence to perform sentence deduplication.
4. It writes its final outputs and some statistics to `../outputs/final_output`.

Here is the help section of the `finalize_data.py` script:

```
$ python finalize_data.py -h
usage: Data Generator [-h] [--working_folder WORKING_FOLDER] [--languages LANGUAGES]
                      [--report_location REPORT_LOCATION]

Generate data from crawls

optional arguments:
  -h, --help            show this help message and exit
  --working_folder WORKING_FOLDER
                        Where is the data
  --languages LANGUAGES
                        Limit to certain languages
  --report_location REPORT_LOCATION
                        Where to create the report.
```

## Example usage

Here is an example call of the script

```
python finalize_data.py  --language hau_Latn,yor_Latn --batch_size 500000
```

and its output shows statistics for each language.

```
Language: hau
script: Latn
words: 829,419
characters: 4,695,920
---
Language: yor
script: Latn
words: 45,549,046
characters: 253,040,908
```

The script generates one text file for each language.

```
$ ls ../outputs/final_output/
final_data_hau_Latn.txt  final_data_yor_Latn.txt  stats.csv  stats.txt
```
